```python
from sympy import symbols, cos, sin, pi
from sympy.vector import CoordSys3D

# Define the coordinate system
N = CoordSys3D('N')

# Define the parameters
u, v = symbols('u v')

# Define the parametric equations for the surface
x = 4 * u - 6 * v 
y = - u  + 2 * v
z = v 

# Define the vector representation of the surface
surface = x * N.i + y * N.j + z * N.k
J = surface.diff(u).cross(surface.diff(v)).magnitude().simplify()

# Display the parametric surface
integrate(integrate(J, (u, -1, 1)), (v, 0, 1))
```
