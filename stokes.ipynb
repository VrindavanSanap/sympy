```python
N = CoordSys3D('N')
x, y, z= N.x, N.y, N.z
x, y, z
```

```python
surface = x**2 * N.i + y * sin(z)*N.j + 3 * z * E**x * N.k
print(surface)
(curl(surface))
```

```python
f = cos(y) - x + 3*x**2 *y + 2 * x*z
```

```python
integrand = f - (sin(y)).diff(y) + (x**2*y - z*x).diff(z)
```

```python
integrate(integrand, x)
```
